```python
# Splitting the target from the other variables

data_X = ks_2018_data_encoded.drop('state_target', axis =1)

data_y= ks_2018_data_encoded.state_target
```

```python
#Splitting the data
from sklearn.model_selection import train_test_split

X_train, X_rem, y_train, y_rem = train_test_split(data_X, data_y, train_size=0.7, random_state=1)

#Where X_train is the training data
```

```python
X_valid, X_test, y_valid, y_test = train_test_split(X_rem, y_rem, test_size=0.5, random_state = 1) 
```

```python
#Spot check analysis using the validation data since we do not know the right model that will fit this data well
# Spot check using two linear models and four non-linear models with KFold cross validation
# Linear models are Logistic regression and Linear Discriminant Analysis
# Non linear models to be used are kNN, NB , RF and SVM

# Imbalanced validation data

#from sklearn.preprocessing import StandardScaler

#scaler = StandardScaler().fit(X_valid)
#rescaled_X_valid = scaler.transform(X_valid)

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot

num_folds = 10
seed = 1

models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
#models.append(('KNN', KNeighborsClassifier()))
#models.append(('RF', RandomForestClassifier()))
#models.append(('NB', GaussianNB()))
#models.append(('SVM', SVC()))

#results = cross_val_score(model,rescaled_X_valid, y_valid, cv=kfold)
# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = KFold(n_splits=num_folds, random_state=seed,shuffle =True)
    cv_results = cross_val_score(model,X_valid, y_valid, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean()*100.0, cv_results.std())
    print(msg)
    
# boxplot algorithm comparison
fig = pyplot.figure()
fig.suptitle('Spot Check model comparison')
ax = fig.add_subplot(111)
pyplot.boxplot(results)
ax.set_xticklabels(names)
pyplot.show()
```

```python
#Balancing the data with SMOTE

from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler


#scaler = StandardScaler().fit(X_valid)
#rescaled_X_valid = scaler.transform(X_valid)

oversample = SMOTE()
X_valid_resam, y_valid_resam = oversample.fit_resample(X_valid, y_valid)

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot

num_folds = 10
seed = 1

models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
#models.append(('KNN', KNeighborsClassifier()))
#models.append(('RF', RandomForestClassifier()))
models.append(('NB', GaussianNB()))
#models.append(('SVM', SVC()))

results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = KFold(n_splits=num_folds, random_state=seed,shuffle =True)
    cv_results = cross_val_score(model,X_valid_resam, y_valid_resam, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean()*100.0, cv_results.std())
    print(msg)
    
# boxplot algorithm comparison
fig = pyplot.figure()
fig.suptitle('Spot Check model comparison')
ax = fig.add_subplot(111)
pyplot.boxplot(results)
ax.set_xticklabels(names)
pyplot.show()
```
